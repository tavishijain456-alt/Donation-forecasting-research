# 📊 Time Series Analysis — STL, Stationarity, ACF/PACF & Descriptive Statistics

## 🔹 1. STL Decomposition (Trend, Seasonality, Residual)

**Concept:**
STL (Seasonal-Trend decomposition using LOESS) is used to break a time series into three components:

* **Trend:** Long-term movement (increasing/decreasing pattern)
* **Seasonality:** Repeating cycles (e.g., yearly patterns)
* **Residual (Noise):** Random variations not explained by trend/seasonality

**Key Parameter:**

* `period`: Number of observations per cycle (e.g., 52 for weekly data with yearly seasonality)

**Why it matters:**

* Helps visually understand structure before modeling
* Confirms whether seasonality exists
* Guides model selection (e.g., ARIMA, Prophet)

---

## 🔹 2. Stationarity & ADF Test

### 📌 What is Stationarity?

A time series is **stationary** if:

* Mean is constant over time
* Variance is constant
* No long-term trend or changing seasonality

**Why stationarity is important:**

* Most models (like ARIMA) assume stationary data
* Non-stationary data leads to poor predictions

---

### 📊 Augmented Dickey-Fuller (ADF) Test

**Purpose:**
Statistically test whether a time series is stationary

**Output:**

* ADF Statistic
* **p-value (most important)**

**Decision Rule:**

* `p < 0.05` → Data is **stationary**
* `p > 0.05` → Data is **non-stationary**

---

### 🔁 Differencing

If data is non-stationary:

* Apply **first-order differencing**

Formula:

```
y_t' = y_t - y_(t-1)
```

**Effect:**

* Removes trend
* Stabilizes mean

**Parameter in ARIMA:**

* Number of times differencing is applied = **d**

---

## 🔹 3. ACF & PACF (Autocorrelation Analysis)

### 📌 Autocorrelation (ACF)

Measures correlation between:

* Current value and past values (lags)

**Used to determine:**

* Moving Average (MA) component → **q**

---

### 📌 Partial Autocorrelation (PACF)

Measures:

* Direct correlation between current value and a lag
* Removes indirect effects of intermediate lags

**Used to determine:**

* AutoRegressive (AR) component → **p**

---

### 📊 Interpreting Plots

| Plot | Use        | Pattern                 |
| ---- | ---------- | ----------------------- |
| ACF  | Find **q** | Sharp cutoff → MA order |
| PACF | Find **p** | Sharp cutoff → AR order |

---

### 📌 Lags

* Lag = previous time step
* Lag 1 → previous week
* Lag k → k weeks ago

---

### 📌 ARIMA Parameters

ARIMA model is defined as:

```
ARIMA(p, d, q)
```

* **p** → from PACF
* **d** → from differencing (ADF test)
* **q** → from ACF

---

## 🔹 4. Descriptive Statistics

### 📊 Key Measures:

* **Mean:** Average value
* **Median:** Middle value
* **Standard Deviation:** Spread of data
* **Minimum:** Lowest value
* **Maximum:** Highest value





---

## 🔹 5. Overall Workflow (Important for Future)

1. **Decompose data** → Understand structure (STL)
2. **Check stationarity** → ADF test
3. **Make stationary** → Differencing
4. **Find parameters** → ACF & PACF
5. **Summarize data** → Descriptive statistics

````python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
def evaluate(actual, predicted):
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mae = mean_absolute_error(actual, predicted)
    mape = np.mean(np.abs((actual-predicted)/actual))*100
    return {'RMSE':round(rmse,2), 'MAE':round(mae,2), 'MAPE':round(mape,2)}
````
